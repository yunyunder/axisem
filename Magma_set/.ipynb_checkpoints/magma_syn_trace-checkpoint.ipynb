```python
def FILE(dir,sta,net,comp):
    path = '/home/yin/axisem/SOLVER/'+dir+'/Data_Postprocessing/SEISMOGRAMS/'
    file = path+sta+'_'+net+'_disp_post_mij_conv0000_'+comp+'.dat'
    data = np.loadtxt(file)
    x,y= data[:,0],data[:,1]*5
    tr_tmp = Trace(y)
    tr_tmp.stats.starttime=starttime
    tr_tmp.stats.delta = x[1]-x[0]
    tr_tmp.stats.channel = comp
    tr_tmp.stats.station = sta
    tr_tmp.stats.network = net
    tr_new = trace_operator(tr_tmp)  #filter 
    return  tr_new

def FILE_trace(dir,sta,net,comp,epi_dis):
    path = '/home/yin/axisem/SOLVER/'+dir+'/Data_Postprocessing/SEISMOGRAMS/'
    file = path+sta+'_'+net+'_disp_post_mij_conv0000_'+comp+'.dat'
    data = np.loadtxt(file)
    x,y= data[:,0],data[:,1]*5
    tr_tmp = Trace(y)
    tr_tmp.stats.starttime=starttime
    tr_tmp.stats.delta = x[1]-x[0]
    tr_tmp.stats.channel = comp
    tr_tmp.stats.station = sta
    tr_tmp.stats.network = net
    tr_tmp.stats.distance = epi_dis
    tr_new = trace_operator(tr_tmp)  #filter 
    return  tr_new


def trace_operator(tr_do):
    fmax=0.1
    fmin=0.01
    tr_do.filter("bandpass", freqmin=fmin, freqmax=fmax,corners=2, zerophase=True)
    tr_do.trim(starttime=starttime,endtime=endtime)
    tr_do.detrend("linear").taper(max_percentage=0.05, type='hann')
    return tr_do


def plot_phase(y):
    for i in range(len(arr_time)):
        ax.axvline(arr_time[i].time,ymin=0.15, ymax=0.85,color='r',linestyle='--')
        ax.text(arr_time[i].time, y.max()*0.87 ,arr_time[i].name,
                 fontsize=12,horizontalalignment='center')
```

```python
def event_info(event):
    if ( event == '20150413_09' ):
        time = '2015-04-13 01:30:28.00'
        elat = 25.286
        elon = 122.399
        edep = 228
    elif (event == '20151128_00'):
        time = '2015-11-28 00:52:23.81'
        elat = 25.1883
        elon = 122.2613
        edep = 219
    elif (event == '20151215_00'):
        time = '2015-12-15 00:26:28.00'
        elat = 25.234
        elon = 122.314
        elon = 217
    elif (event == '20151216_10'):
        time = '2015-12-16 10:30:58.00'
        elat = 25.223
        elon = 122.13
        edep = 218
    elif (event == '20180117_05'):
        time = '2018-01-17 05:59:11.04'
        elat = 25.0588
        elon = 121.7181
        edep = 129
    elif (event == '20190312_20'):
        time = '2019-03-12 20:19:15.39'
        elat = 25.0176
        elon = 121.8555
        edep = 131
    return time ,elat ,elon, edep

```

```python
# write SAC with header details 

from obspy.io.sac import SACTrace
from obspy.taup import TauPyModel
import numpy as np
import matplotlib.pyplot as plt
from obspy import read, Stream, UTCDateTime, Trace
from obspy.geodetics import degrees2kilometers
from matplotlib.transforms import blended_transform_factory
import os




dir_list = np.genfromtxt('prem_test',dtype=str)

for dir in dir_list:
    print(dir)
    
    time, elat, elon, edep = event_info( dir[8:19])
    print('%s EVENT'%(dir[8:19]))
    starttime=UTCDateTime(time)
    endtime=starttime+150
    
    
    sac_dir = 'sac_'+dir
    os.mkdir(sac_dir)
    os.chdir(sac_dir)
    
    
    sta_file = '/home/yin/axisem/SOLVER/'+dir+'/Data_Postprocessing/receiver_gll_locs_mrr.dat'
    sta_info = np.genfromtxt(sta_file, skip_header=1,dtype=str)
    sta_name, epi_dis= sta_info[:,0], sta_info[:,4].astype(float)
    sta_lat,sta_lon = sta_info[:,2].astype(float), sta_info[:,3].astype(float)
    
    num = []
    epi_dis.argsort()[0]
    model = TauPyModel(model='prem')

    for i in range(0,len(sta_info),30): #3
        num.append( epi_dis.argsort()[i] )

    st = Stream()
    for i in num:
        #sta = sta_name[i]    #epi = epi_dis[i] #data = np.loadtxt(sta+'.MXE.sem.ascii')
        tr_tmp = Trace()
        epi_m = degrees2kilometers(epi_dis[i])*1000
        tr_tmp = FILE_trace(dir, sta_name[i].split('_')[0], sta_name[i].split('_')[1], 'Z',epi_m)
        tr_tmp_r = FILE_trace(dir, sta_name[i].split('_')[0], sta_name[i].split('_')[1], 'R',epi_m)
        tr_tmp_t = FILE_trace(dir, sta_name[i].split('_')[0], sta_name[i].split('_')[1], 'T',epi_m)
        st += tr_tmp.copy()
        arr_time = model.get_travel_times(source_depth_in_km=edep,
                                      distance_in_degree=epi_dis[i],
                                      phase_list=['p','s'])
        #arr_time = model.get_travel_times(source_depth_in_km=219.,
        #                              distance_in_degree=epi_dis[i],
        #                              phase_list=['p'])
        #print(arr_time[1])
        #time = arr_time[0].time
        for tmp_comp in (tr_tmp_r , tr_tmp_t, tr_tmp):
            sac = SACTrace.from_obspy_trace(tmp_comp)
            sac.evla = elat
            sac.evlo = elon
            sac.evdp = edep
            sac.stla = sta_lat[i]
            sac.stlo = sta_lon[i]
            sac.stdp = 0.
            sac.stel = 0.
            sac.o = starttime
            sac.iztype = 'io' 
            sac.lcalda = True 
            sac.t0 = arr_time[0].time
            sac.t2 = arr_time[0].time
            sac.kt0 = arr_time[0].name
            sac.t9 = arr_time[1].time
            sac.kt9 = arr_time[1].name
    
            sac_name = ('%s.%s.%s.c2.SAC')%(sac.kstnm, sac.knetwk, sac.kcmpnm)
            #print(sac_name)
            sac.write(sac_name)
    os.chdir('../')

print('Done!')
```
